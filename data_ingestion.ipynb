# Import csv data and quick analysis

- for all data county feature = ID code of that country 
- the mapping of all the countries is in the json file
- product_type : 0: Combined, 1: Fixed, 2: General service, 3: Spot
- data_block_id - All rows sharing the same data_block_id will be available at the same forecast time. This is a function of what information is available when forecasts are actually made, at 11 AM each morning. For example, if the forecast weather data_block_id for predictins made on October 31st is 100 then the historic weather data_block_id for October 31st will be 101 as the historic weather data is only actually available the next day.

Intuition for merging weather history data : 
- First we take each longitude and latitude and map it into the country it's from
- Next we take each hour and calculate the mean of it in each country
- Merge the weather_history with the train_data on country and date_time

```python
weather_history_df
```

```python
# Import the weather to country mapping

weather_station = pd.read_csv("weather_station_to_county_mapping.csv")
```

```python
# Drop unnecesarry columns

weather_station.drop("county_name", axis=1, inplace=True)
```

```python
# Fill nulls with ffill because the data is ordered from low to high, use after that backfill in order to fill the first values

weather_station.fillna(method="ffill", inplace=True)
weather_station.fillna(method="backfill", inplace=True)
```

```python
# Change country's type

weather_station["county"] = (weather_station["county"]).astype(int)
```

```python
# Add country column in weather_history using the map in weather_station

for row in weather_station.index:
    weather_history_df.loc[
        weather_history_df[
            np.isclose(
                weather_history_df["longitude"], weather_station.iloc[row]["longitude"]
            )
            & np.isclose(
                weather_history_df["latitude"], weather_station.iloc[row]["latitude"]
            )
        ].index,
        "county",
    ] = weather_station.iloc[row]["county"]
```

```python
# Drop longitude and latitude because we calculate weather condition based on country

weather_history_df.drop(["latitude", "longitude"], inplace=True, axis=1)
```

```python
# Drop unnecesarry and highly correlated columns

weather_history_df.drop(
    [
        "dewpoint",
        "snowfall",
        "cloudcover_total",
        "diffuse_radiation",
        "shortwave_radiation",
        "data_block_id",
        "surface_pressure",
    ],
    inplace=True,
    axis=1,
)
```

```python
# Calculate the mean of the temperature, cloud_coverage... etc for each hour and country

weather_history_average_df = weather_history_df.groupby(["datetime", "county"]).mean()
```

```python
# Reset indexes and convert into date to merge on it

weather_history_average_df.reset_index(inplace=True)
weather_history_average_df["datetime"] = pd.to_datetime(
    weather_history_average_df["datetime"]
)
```

```python
# Since all weather conditions were known before 2023-05-30 10:00:00 we will use weather forecast to fill remaining rows

weather_forecast_df = weather_forecast_df[
    (
        # Select the first origin_datetime after the last known weather conditions
        weather_forecast_df["origin_datetime"]
        == next(
            x
            for x in weather_forecast_df["origin_datetime"][::-1]
            if x < str(weather_history_average_df["datetime"].iloc[-1])[:10]
        )
    )
    & (
        # Select the only forecast dates that we need
        weather_forecast_df["forecast_datetime"]
        < str(train_df_merged["datetime"].iloc[-1] + pd.Timedelta(1, unit="h"))[:10]
    )
    & (
        # Select the forecast dates that comes after the last nown weather conditions
        weather_forecast_df["forecast_datetime"]
        > str(
            weather_history_average_df["datetime"].iloc[-1] + pd.Timedelta(1, unit="h")
        )
    )
]
```

```python
# Calculate wind speed and direction

weather_forecast_df["winddirection_10m"] = 180 + (180 / np.pi) * np.arctan2(
    weather_forecast_df["10_metre_u_wind_component"],
    weather_forecast_df["10_metre_v_wind_component"],
)
weather_forecast_df["windspeed_10m"] = np.sqrt(
    np.power(weather_forecast_df["10_metre_v_wind_component"], 2)
    + np.power(weather_forecast_df["10_metre_u_wind_component"], 2)
)
weather_forecast_df["total_precipitation"] = (
    weather_forecast_df["total_precipitation"] * 1000
)
```

# Setup ML Pipelines

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import SGDRegressor
```

```python
model = SGDRegressor(random_state=1547)
```

```python
def divide_batches(X, y, batch_size):
    batch_list = []

    while len(X != 0):
        batch = []
        batch_y = []

        batch = X[:batch_size]
        X = X[batch_size:]

        batch_y = y[:batch_size]
        y = y[batch_size:]

        batch_list.append([batch, batch_y])

    return batch_list
```

```python
batches = divide_batches(X=X_train, y=y_train, batch_size=10000)
```

```python
X
```

```python
def model_fit(model, batches_list, number_of_batches):
    for i in range(number_of_batches):
        model.partial_fit(batches_list[i][0], batches_list[i][1])
    return model
```

```python
trained_model = model_fit(model=model, batches_list=batches, number_of_batches=10)
```
